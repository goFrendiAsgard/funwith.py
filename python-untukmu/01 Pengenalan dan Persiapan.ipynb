# Pengenalan Komputer

Pada awalnya komputer adalah sebuah alat yang digunakan untuk melakukan kegiatan komputasi. Namun dalam perkembangan selanjutnya komputer bukan lagi sekedar alat untuk berkomunikasi. Komputer telah menjadi bagian hidup banyak orang. Kita dapat menggunakan komputer untuk bermain game, mendengarkan musik, menonton film, chatting, browsing, belajar, bahkan saling mem-follow dan mencerca di jejaring sosial semacam facebook/twitter.

Wujud komputer pun berubah seiring waktu. Jika dahulu kala, sebuah komputer memiliki ukuran sebesar sebuah ruangan, sekarang komputer telah menjadi semakin porTabel, sehingga bisa diletakkan di pangkuan anda (laptop), masuk ke dalam saku celana anda (smartphone), bahkan bisa anda pakai (smart-watch dan google-glass). Untuk keperluan yang lebih khusus, bahkan telah diciptakan komputer dengan ukuran mikroskopik.

![](files/resource/perkembangan-komputer.png)

Perubahan ini tak terlepas dari peranan teknologi. Pada awalnya, diguanakan *tabung hampa* untuk mengubah state data dalam komputer. Tabung hampa tersebut berukuran cukup besar, rapuh, dan lambat karena adanya batasan-batasan mekanik. Namun kemudian, ditemukan apa yang disebut sebagai transistor. Transistor memanfaatkan kepingan semikonduktor, sehingga bisa terlepas dari batasan-batasan mekanik. Lama kelamaan, ukuran transistor pun semakin kecil, hingga mencapai beberapa puluh molekul saja. Dengan ini, para ilmuwan telah bisa menciptakan apa yang disebut sebagai *nano-computer*.

Dibandingkan dengan beberapa dekade yang lalu, kini komputer bukanlah alat elit yang hanya bisa diakses oleh profesor-profesor di perguruan tinggi ternama. Komputer telah menjadi bagian dari kehidupan sehari-hari kita.
Dalam menggunakan komputer, biasanya anda akan mengakses aplikasi (program komputer). Ada berbagai macam aplikasi yang sangat mudah digunakan. Microsoft word dan Libre-office word misalnya. Dengan mudah anda bisa mengetikkan dokumen, memberikan nomor halaman, memberikan format-format yang menawan serta mencetaknya pada kertas. Adobe photoshop dan Gimp, memungkinkan anda mengedit foto dengan mudah, menghilangkan kerutan wajah dan jerawat, bahkan membuat kulit anda terlihat lebih cerah tanpa perlu menggunakan krim pemutih wajah.

Kebanyakan aplikasi komputer yang ada memang sengaja dibuat untuk kemudahan pengguna. Aplikasi-aplikasi tersebut dibuat oleh sekelompok orang yang disebut sebagai programmer. Para pembuat program.

Untuk membuat program, anda juga membutuhkan aplikasi pembuat program. Ada banyak aplikasi yang digunakan untuk tujuan itu, misalnya Microsoft-visual-studio, Delphi, Java, C dan sebagainya. Aplikasi-aplikasi ini dibuat untuk programmer, sehingga terkadang penggunaannya tidak semudah penggunaan aplikasi-aplikasi yang ditujukan untuk pengguna awam.

Seorang programmer harus bisa berpikir secara logis, formal dan sederhana. Pemikiran sang programmer perlu diterjemahkan ke dalam bentuk set perintah untuk komputer yand ditulis dalam bahasa pemrograman tertentu. Pemikiran secara logis, formal, dan sederhana tersebut disebut algoritma.

# Algoritma

Algoritma adalah langkah demi langkah yang dilakukan untuk menyelesaikan masalah. Misalnya anda
sedang kelaparan, dan kebetulan ada rombong bakso yang lewat di depan rumah/kos anda. Dalam
kasus ini masalahnya sudah jelas. Anda lapar. Solusinya juga seharusnya sudah jelas, anda harus
makan. Pertanyaannya adalah bagaimana supaya solusi itu bisa tercapai? Anda tidak menghampiri
rombong bakso tersebut, mengambil beberapa gorengan, dan pergi begitu saja. Ada langkah-langkah
yang harus anda lakukan. Coba perhatikan langkah-langkah berikut:

1. Periksa uang di saku anda
2. Jika uang di saku anda kurang dari Rp 10.000,- kembali ke rumah, ambil uang. Jika uang di saku anda lebih besar atau sama dengan Rp 10.000,- maka menuju ke langkah 4, selain itu menuju ke langkah 3.
3. Pinjam uang ke teman atau minta uang ke orang tua, sampai jumlah total uang yang anda bawa mencapai nominal Rp 10.000,-
4. Pergi ke rombong bakso, beli apa yang anda mau (jangan lebih dari Rp 10.000,-)
5. Makan apa yang sudah anda beli
6. Bayar apa yang sudah anda beli

Cukup mudah dipahami bukan? Tentu saja, agak sulit untuk membuat algoritma sempurna yang bisa
menangani semua kasus dalam kehidupan nyata.

Contoh Algoritma dalam Bahasa Indonesia
=======================================

Tujuan algoritma adalah menyelesaikan masalah. Ada berbagai macam masalah yang kerap kita hadapi, mulai dari masalah serius sampai masalah yang tidak terlalu serius.

Semisal, saat ini anda ingin mengasah intuisi anda untuk menebak angka yang akan muncul dari sebuah lemparan dadu, namun anda tidak memiliki dadu. Anda hanya memiliki sebuah laptop di pangkuan anda. Bagaimana anda dapat memenuhi keinginan anda bermain dadu?

Anda bisa memerintahkan komputer untuk membuat bilangan acak dan berinteraksi dengan anda sedemikian rupa sehingga anda bisa menebak bilangan acak yang dibuat oleh komputer.

Berikut adalah langkah yang harus dilakukan komputer:

1. Biarkan komputer mengacak sebuah angka antara 1 sampai 6
2. Munculkan pemberitahuan ke pengguna supaya memasukkan tebakan
3. Baca tebakan pengguna
4. Jika tebakan pengguna sama dengan angka yang ditebak komputer, menuju langkah 5, jika tidak menuju ke langkah 6
5. Munculkan pesan "Anda benar"
6. Munculkan pesan "Anda salah"

Anda baru saja melihat sebuah algoritma sederhana. Algoritma tersebut ditulis dalam bahasa Indonesia.

Tanpa bermaksud menjelek-jelekkan bahasa Indonesia (dan bahasa manusia secara umum), penulisan dalam bentuk ini kerap kali mengundang kesalah-pahaman. Bahasa manusia kerapkali mengundang interpretasi yang berlainan tergantung dari budaya dan kebiasaan orang yang membacanya. Coba perhatikan lagi pernyataan nomor `1`
```
Biarkan komputer mengacak sebuah angka antara 1 sampai 6
```

Bagaimana anda menginterpretasikan kata __antara__ dalam statement tersebut?

Apakah komputer boleh menghasilkan angka `2,5`, `3,4` atau `5,1`? Atau apakah hanya bilangan bulat saja yang boleh dihasilkan? Apakah angka `1` dan `6` sendiri termasuk dalam jangkauan `antara` yang dimaksud? Atau komputer hanya boleh menghasilkan angka `2`, `3`, `4`, dan `5` saja, tanpa `1` dan `6`?

Karena ambuguitas yang ada dalam bahasa manusia (yang sehari-hari kita pakai), maka akan lebih baik jika kita bisa meminimalisasi penggunaan istilah-istilah ambigu tersebut. Kita membutuhkan suatu bentuk yang lebih formal.

Selain ambiguitas, masalah lain yang muncul dalam penulisan dengan gaya seperti ini adalah kesulitan untuk membaca alur proses secara umum. Anda harus membaca statement no 4 dengan teliti untuk memahami bahwa langkah `6` tidak dijalankan setelah langkah `5`.

Secara intuitif, kita akan cenderung mengasumsikan bahwa langkah 2 dijalankan setelah langkah 1, langkah 3 dijalankan setelah langkah 2, dan seterusnya. Namun pada langkah nomor `4` tertulis pernyataan berikut:

```
Jika tebakan pengguna sama dengan angka yang ditebak komputer, menuju langkah 5, jika tidak menuju ke langkah 6
```

Artinya, dari langkah `4` kita bisa menuju langkah `5` ataupun `6` tergantung dari benar tidaknya tebakan kita.

# Bahasa Pemrograman

Selain mampu ber-algoritma, seorang programmer juga harus memiliki kemampuan untuk menuliskan algoritmanya ke dalam bahasa pemrograman yang dimengerti oleh komputer.

Ada berbagai macam bahasa pemrograman. Anda mungkin pernah mendengar sepintas tentang salah satu atau beberapa bahasa pemrograman populer berikut:

* C
* Pascal
* Java
* PHP
* Javascript
* Python

Daftar di atas hanyalah sebagian kecil dari bahasa pemrograman yang ada. Masih ada bahasa pemrograman lain seperti `haskell`, `fortran`, `scala`, `perl`, `ruby`, `dart`, `go`, `matlab`, `R`, `assembler`, `brainfuck` dan sebagainya. Ada ratusan bahasa pemrograman yang saat ini beredar. Masing masin memiliki karakteristik sendiri. `Matlab` misalnya, sering digunakan untuk kalkulasi matriks, `R` untuk statistik, `assembler` untuk pemrograman mikro-kontroler, dan `brainfuck` untuk keperluan iseng.

Berikut, disajikan sedikit penjelasan tentang sebagian bahasa pemrograman yang ada:

## C

C adalah salah satu bahasa pemrograman tertua. Hampir semua programmer mengenal bahasa pemrograman ini.
Untuk membuat program dalam bahasa C, dibutuhkan pengetahuan secara mendalam tentang pengalamatan dan pengalokasian memori dalam komputer. Bahasa ini cukup sulit dipelajari bagi pemula, namun sangat powerful. Sistem operasi seperti windows dan Linux pun konon ditulis dalam bahasa C.

Script program dalam bahasa C perlu di-compile ke dalam bahasa mesin untuk bisa dijalankan.

Contoh cuplikan program dalam bahasa C:
```
int main(){
    int a = 5;
    int b = 7;
    printf("%d", a+b);
    return 1;
}
```

## Pascal

Pascal merupakan bahasa pemrograman yang cukup diminati karena kejelasan syntax nya.

Sama halnya seperti C, script program dalam bahasa pascal perlu di-compile ke dalam bahasa mesin agar bisa dijalankan.

Contoh cuplikan program dalam bahasa Pascal:

```
var
  a : integer;
  b : integer;
begin
  a := 5;
  b := 7;
  write(a+b);
end.
```

## Java

Java merupakan salah satu bahasa pemrograman yang banyak dipakai dalam dunia kerja. Java terkenal karena paradigma OOP nya yang sangat kuat. Java bisa berjalan di berbagai sistem operasi yang berbeda karena sifatnya yang semi-interpreted.

Script program java akan dikompilasi menjadi byte-code. Byte-code ini bisa berjalan di setiap komputer yang memiliki JVM (Java virtual machine). Setiap sistem operasi memiliki JVM yang berbeda untuk menerjemahkan byte-code ke dalam bahasa mesin yang sesuai dengan sistem operasi terkait.

Dibandingkan C & Pascal, program yang ditulis dalam bahasa java sedikit lebih lambat.

Berikut adalah contoh program dalam bahasa java:

```
public class Program{
  public static void main(String[] args){
    int a = 5;
    int b = 7;
    System.out.print(a+b);
  }
}
```

__Catatan:__ OOP (Object Oriented Programming) adalah paradigma pemrograman berbasis obyek. Dalam OOP data dianggap sebagai obyek yang memiliki sekumpulan `properti` dan `metode`. Semisal, obyek `kucing` memiliki beberapa properti seperti `nama`, `jumlah kaki`, dan `warna bulu`. Obyek `kucing` juga memiliki beberapa `metode` seperti `berlari` dan `bermain`.

## PHP

PHP adalah bahasa pemrograman yang banyak digunakan dalam dunia pemrograman web. Script PHP tidak perlu di-compile karena sifatnya yang interpreted, dan karenanya lebih lambat jika dibandingkan C, Pascal, dan Java.

Berikut adalah contoh program dalam bahasa PHP:

```
$a = 5;
$b = 7;
echo $a+$b
```

## Javascript

Javascript adalah bahasa pemrograman yang umum dipakai dalam dunia pemrograman web. Jika PHP berjalan di sisi server, maka javascript pada awalnya berjalan di sisi client. Adapun demikian, kini telah ada beberapa framework seperti node-js yang memungkinkan kita untuk menulis javascript di sisi server.

Sama halnya seperti PHP, javascript juga tidak perlu di-compile.

Berikut adalah contoh program dalam javascript:

```
var a = 5;
var b = 7;
alert(a+b);
```

## Python

Sama halnya seperti PHP dan Javascript, Python juga adalah bahasa pemrograman interpreted. Bahasa ini merupakan salah satu bahasa yang sangat disarankan bagi pemula. Walau demikian, beberapa computer-scientist termasuk Peter Norvig dan Philip Guo memilih Python karena kesederhanaan syntax nya. Bahasa ini banyak digunakan di lingkungan akademis.

Berikut adalah contoh program dalam bahasa python:
```
a, b = 5, 7
print(a+b)
```

Dalam buku ini akan digunakan bahasa Python. Tapi tetap saja tidak ada salahnya anda mempelajari bahasa pemrograman yang lain.

Kelebihan Python
----------------

Berikut adalah beberapa kelebihan Python:

* __Free dan open source__ 

    Anda tidak perlu membayar sepeserpun untuk menggunakan Python. Sifatnya yang open-source membuat banyak programmer menyediakan module-module secara gratis pula.
    
    
* __Mudah dibaca__ 

    Python mirip dengan bahasa Inggris sederhana. Untuk membuat program dalam bahasa Python, anda tidak harus menuliskan berbagai simbol seperti `;` atau `$`. Karenanya kode program dalam bahasa Python menjadi lebih mudah dibaca dan dipahami.
    
    
* __Mudah dipelajari__

    Python tergolong bahasa yang sangat mudah dipelajari. Beberapa tutorial dasar pemrograman di internetpun menggunakan Python.
    

* __Multi purpose, multi paradigm__

    Python dibuat untuk tujuan umum. Bahasa pemrograman ini mengadopsi banyak paradigma (modular ataupun OOP), namun tidak memaksa penggunanya untuk terikat pada satu paradigma tertentu
    

* __Komunitas yang kuat__

    Jika menemukan permasalahan, anda dapat mengunjungi stackoverflow dan mendapatkan jawaban dalam waktu singkat.
    

* __Package System__

    Python memiliki package system yang disebut `pip`. Sama halnya seperti `gem` di ruby, `composer` di PHP, ataupun `npm` di node-js. Dengan adanya package system ini, anda tidak perlu membuat keseluruhan program dari awal. Anda bisa menggunakan flask atau bottle untuk membuat aplikasi web, anda juga bisa menggunakan matplotlib atau numpy untuk kepentungan scientific. Hampir segala sesuatu telah disediakan di sini.
    
    
Kekurangan Python
-----------------

* __Tidak mendukung functional programming secara penuh__

    Pemrograman functional bukanlah paradigma yang umum digunakan oleh sebagian besar programmer. Functional Programming umumnya digunakan oleh matematikawan. Jika anda ingin mempelajari functional-programming, Python bukanlah bahasa terbaik untuk itu.
    
    __Catatan:__ Functional Programming adalah paradigma pemrograman berbasis fungsi. Berbeda dengan pemrograman `imperative` seperti C atau Java, dalam `functional programming` tidak boleh ada `state` atau kondisi yang tersimpan dalam variabel. Dalam `functional programming`, program tidak berjalan secara urut dari atas ke bawah, melainkan berdasarkan mana yang perlu dieksekusi terlebih dahulu. Contoh bahasa pemrograman yang mendukung penuh `functional programming` adalah haskell dan scala.
    
    
* __Relatif lambat jika dibandingkan dengan C__
    
    Walaupun script Python tampak jauh lebih sederhana dari C, namun komputer membutuhkan waktu yang lebih lama untuk mengeksekusi program yang ditulis menggunakan Python. Jika anda membutuhkan program dengan kecepatan yang sangat tinggi, anda mungkin akan harus menggunakan bahasa C, atau setidaknya menggunakan compiler Python yang sudah dimodifikasi seperti Cython atau Pypy.

# Persiapan

Python tidak terinstall secara default di windows. Jika anda menggunakan windows, maka sebaiknya anda mendownload anaconda. Anaconda adalah paket installasi Python yang cukup lengkap. Di dalamnya sudah ada library-library yang umum dipakai seperti matplotlib dan numpy. Untuk mendownload anaconda, silahkan kunjungi website berikut: [http://continuum.io/downloads](http://continuum.io/downloads).

Python biasanya telah terinstall secara default pada komputer mac dan linux. Jika anda membutuhkan library-library lain, dan menginginkan proses instalasi yang mudah, anda dapat mendownload anaconda untuk linux atau mac di alamat yang sama. Anda juga bisa menggunakan pip untuk menginstal package secara manual.

Pengguna ubuntu bisa menjalankan perintah berikut di bash-shell untuk menginstal IPython (interface interaktif Python):

```
$ sudo apt-get install python-pip python-dev build-essential 
$ sudo pip install --upgrade pip
$ sudo pip install --upgrade ipython[all]
```

Untuk menginstal package lain secara manual (misalnya matplotlib atau numpy) anda bisa menjalankan perintah berikut:

```
$ sudo pip install matplotlib
$ sudo pip install numpy
```

Kini anda sudah siap untuk memulai petualangan dengan Python.

# Menjalankan Script Python

Ada tiga cara untuk menjalankan script Python:

* Menyimpan script Python dalam file dengan ekstensi `.py`, dan menjalankannya dengan perintah `python namafile.py`

    Cara ini adalah cara yang paling umum digunakan jika anda ingin membuat program secara serius dan menyimpan script anda secara permanen.
    

* Menggunakan mode interaktif (ipython) dengan cara menjalankan perintah `ipython`

    Cara ini biasa digunakan untuk keperluan prototyping.
    

* Menggunakan mode interaktif web (ipython notebook) dengan cara menjalankan perintah `ipython notebook`

    Cara ini biasa digunakan untuk keperluan prototyping dan penelitian. Dalam mode ini, selain menulis script Python, anda juga bisa menulis catatan dan persamaan matematika. Buku yang sedang anda baca ini juga ditulis menggunakan Ipython notebook.
    

Cara pertama: menjalankan script Python yang tersimpan dalam file
=================================================================
Untuk melakukan ini, anda membutuhkan sebuah text-editor seperti notepad, notepad++, gedit, sublimetext atau IDLE.
Ikuti langkah-langkah berikut:

* Buat sebuah file `hello.py`
* Buka file tersebut, dan tuliskan script berikut:

```
print("Hello world")
```
    
* Jalankan dengan membuka command prompt dan mengetikkan perintah berikut untuk mac atau linux:
    
```
python lokasi\penyimpanan\hello.py
```

* Atau jalankan dengan membuka command prompt dan mengetikkan perintah berikut untuk windows:
    
```
python lokasi/penyimpanan/hello.py
```

===================

Cara ketiga: IPython Notebook
=============================

* Buka command prompt dan ketikkan perintah `ipython notebook`
